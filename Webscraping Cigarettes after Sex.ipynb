```python
import requests
from bs4 import BeautifulSoup
```

```python
url = 'https://www.lyrics.com/artist/Cigarettes-After-Sex/3270020'
```

```python
#in order to not get blocked, mimicking website
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}

#get html content 
response = requests.get (url, headers=headers)

#get html contained w code below
response.text
```

```python
# check status code

response.status_code
```

```python
#transforms response object into a soup object- parses through the document, knows the exact structure of html 

soup = BeautifulSoup(response.text, 'html.parser') #put parser in as a string
```

```python
soup
```

```python
# find where the poem title and link is in the html page (inspect: class_ item-description-title) - can see belongs under span tag and has the class description as written below

soup.find('div').find('li').a['href']
#soup.find(class_='tal qx').a['href']
#this is one song
```

```python
# use find_all (this is to extract all the titles)-> all occurrences of where particular class is defined

cas_lyrics = soup.find_all(class_='tal qx') 
```

```python
cas_lyrics
```

```python
titles = []

for title in cas_lyrics:
    titles.append(title.a.text)
```

```python
titles
```

```python
# Extract all song links
links = []
songlinks = soup.find_all('td', attrs={'class' : 'tal qx'})
for link in songlinks:
    links.append("https://www.lyrics.com"+link.a['href'])
links
```

```python
# Put the info of title and link in dataframe
import pandas as pd
df=pd.DataFrame({'title': titles, 'links': links})
df
```
